```python
import pandas as pd
```

```python
df_adc_trials = pd.read_csv("PanCancer_ADC_trials_supp5.csv", header=1)
```

```python
from pyensembl import ensembl_grch38
```

```python
df_adc_trials[df_adc_trials.Symbols.isnull()]
```

```python
gene_ids = []
for names in df_adc_trials.Symbols:
    print(names)
    sub_names = names.split(";")
    sub_ids = []
    for name in sub_names:
        
        print(name)
        genes = ensembl_grch38.genes_by_name(name)
        print(genes)
        gene_id = genes[0].id
        sub_ids.append(gene_id)
    gene_ids.append(";".join(sub_ids))
df_adc_trials["Ensembl_Gene_IDs"] = gene_ids
df_adc_trials.to_csv("ADC-trials.csv", index=False)
```
